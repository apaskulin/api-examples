# Retrieving free knowledge

Let's search for a page.

```python
import requests

url = 'https://en.wikipedia.org/w/rest.php/coredev/v0/search/page'
headers = {
    'User-Agent': 'MediaWiki REST API docs examples/0.1 (https://meta.wikimedia.org/wiki/User:APaskulin_(WMF))'
}
response = requests.get(url, headers=headers, params={'q': 'dog'},)

import json
response = json.loads(response.text)
print('\nSearch results:')
for page in response['pages']:
    print(page['title'])
```

Now let's see the available languages for the page: Dog. 

```python
url = 'https://en.wikipedia.org/w/rest.php/coredev/v0/page/Dog/links/language'
response = requests.get(url, headers=headers)
response = json.loads(response.text)
print('\nLanguages:')
for language in response:
    print(language['name'])
```
